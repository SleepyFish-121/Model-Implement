```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_context("notebook")
```

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.get('data')[:,:2]
y = iris.get('target')
```

```python
from Library.Neighbors import KNeighborsClassifier
from Library.Score import r2_score
from sklearn.model_selection import train_test_split

train_X,test_X,train_y,test_y = train_test_split(X,y,test_size=0.3,random_state=5)
classifier=KNeighborsClassifier(n_neighbors=3).fit(train_X,train_y)
pred_y=classifier.predict(test_X)
r2_score(test_y,pred_y)
```

```python
classifier=KNeighborsClassifier(n_neighbors=3).fit(X,y)
```

```python
x_min=np.min(X[:,0])
x_max=np.max(X[:,0])
y_min=np.min(X[:,1])
y_max=np.max(X[:,1])

xx,yy=np.meshgrid(np.arange(x_min,x_max,0.02), np.arange(y_min,y_max,0.02))
labels=classifier.predict(np.c_[xx.ravel(), yy.ravel()])
plt.contourf(xx, yy,labels.reshape(xx.shape), alpha=0.4)

plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```
