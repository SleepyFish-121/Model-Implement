```python
n = 200  
scale = 50  
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_context("poster")
```

```python
from Library.LinearModel import *
```

# Generate data

```python
np.random.seed()
X = np.array(range(0, n)).reshape(-1, 1)
b = np.random.random((1, 1)) / np.random.random((1, 1)) - np.random.random((1, 1)) / np.random.random((1, 1))
b0 = np.random.random((1)) / np.random.random((1)) - np.random.random((1)) / np.random.random((1))
noise = np.random.normal(scale=scale, size=n)
y = np.matmul(X, b) + b0
observed_y= y + noise.reshape(-1, 1)
```

# Testing liner regression model

```python
plt.figure(figsize=(16, 8))
plt.plot(X, y, color='black', label='True Value')
plt.scatter(X, observed_y, color='red', label='Observation Value')
model = LinearRegreesion().fit(X, observed_y)

y = model.predict(X)
plt.plot(X, y, color='green', label='Regression Model')
plt.legend()
plt.title("Comparison between regression model and true value")
plt.show()
```

# Test ridge regression model

```python
plt.figure(figsize=(16, 8))
plt.plot(X, y, color='black', label='True Value')
plt.scatter(X, observed_y, color='red', label='Observation Value')
model = Ridge(alpha=.5).fit(X, y)

y = model.predict(X)
plt.plot(X, y, color='green', label='Regression Model')
plt.legend()
plt.title("Comparison between regression model and true value")
plt.show()
```
