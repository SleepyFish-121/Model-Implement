# Testing liner model with 1 unknown coefficients and visulization the result

```python
# n>p
n = 200  # This is the number of  unknown coefficients(including constant).
scale = 50  # This is the std of noise
np.random.seed()

X = np.array(range(0, n)).reshape(-1, 1)
b = np.random.random((1, 1)) / np.random.random((1, 1)) - np.random.random((1, 1)) / np.random.random((1, 1))
b0 = np.random.random((1)) / np.random.random((1)) - np.random.random((1)) / np.random.random((1))
noise = np.random.normal(scale=scale, size=n)
y = np.matmul(X, b) + b0
plt.figure(figsize=(16, 8))
plt.plot(X, y, color='black', label='True Value')

y = y + noise.reshape(-1, 1)
plt.scatter(X, y, color='red', label='Observation Value')
model = LinearRegreesion().fit(X, y)

y = model.predict(X)
plt.plot(X, y, color='green', label='Regression Model')
plt.legend()
plt.title("Comparison between regression model and true value")
plt.show()
```
